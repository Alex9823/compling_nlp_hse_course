```python
ent1 = []
sents = []
ent2 = []
left = []
right = []
middle = []
target = []

for key in data:
    for inst in data[key]:
        tokens = inst['tokens']
        sents.append(tokens)
        ent1.append(' '.join([tokens[i] for i in inst['h'][2][0]]))
        ent2.append(' '.join([tokens[i] for i in inst['t'][2][0]]))
        order = True if inst['h'][2][0][0] < inst['t'][2][0][0] else False
        
        if order:
            i = inst['h'][2][0][0]
            il = inst['h'][2][0][-1]
            j = inst['t'][2][0][0]
            jl = inst['t'][2][0][-1]
            
            l = tokens[max(0, i-3):i]
            left.append((['<START>']*(3-len(l))) + l)
            r = tokens[jl:jl+3]
            right.append(r + (['<END>']*(3-len(r))))
            
            m = tokens[il:j]
            middle.append(m)
            
            target.append(key)
        else:
            i = inst['t'][2][0][0]
            il = inst['t'][2][0][-1]
            j = inst['h'][2][0][0]
            jl = inst['h'][2][0][-1]
            
            l = tokens[max(0, i-3):i]
            left.append((['<START>']*(3-len(l))) + l)
            r = tokens[jl:jl+3]
            right.append(r + (['<END>']*(3-len(r))))
            
            m = tokens[il:j]
            middle.append(m)
            
            target.append(key)
        
```
