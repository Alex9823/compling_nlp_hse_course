```python
sents_with_ents = []

for file in files:
    js = json.load(open('news/'+file))
    text = js['text']
    doc = nlp(text)
    
    
    for sent in doc.sents:
        sent = sent.as_doc()
        pers = []
        orgs = []
        for entity in sent.ents:
            if entity.label_ == 'PERSON':
                pers.append(entity)
            elif entity.label_ == 'ORG':
                orgs.append(entity)
        if pers and orgs:
            sents_with_ents.append((sent, pers, orgs))
                
                
```

```python
s = sents_with_ents
```

```python
lefts = []
ents1 = []
ents2 = []
rights = []
middles = []
orders = []
sents = []
tags = []
for sent, pers, orgs in sents_with_ents:
    for per in pers:
        pairs = [[per, org] for org in orgs]
        sent=list(sent)
        for ent1, ent2 in pairs:
            sents.append(' '.join([str(w) for w in sent]))
            if ent1.start < ent2.start:
#                 order.append(1)
                
                lefts.append(sent[max(0, ent1.start-3):ent1.start])
                rights.append(sent[ent2.end:ent2.end+3])
                middles.append(sent[ent1.end:ent2.start])
                ents1.append(ent1)
                ents2.append(ent2)
                tags.append([ent1.label_, ent2.label_])
            else:
#                 order.append(0)
                sent=list(sent)
                lefts.append(sent[max(0, ent2.start-3):ent2.start])
                rights.append(sent[ent1.end:ent1.end+3])
                middles.append(sent[ent2.end:ent1.start])
                ents1.append(ent2)
                ents2.append(ent1)
                tags.append([ent2.label_, ent1.label_])
            
```

```python
def pad(seqs, n=3, left=True, padding='<PAD>'):
    padded = []
    for seq in seqs:
        if left:
            padded.append(([padding]*(n-len(seq)))+seq)
        else:
            padded.append(seq+([padding]*(n-len(seq))))
    return padded
```

```python
lefts = pad(lefts, padding='<START>')
rights = pad(rights, left=False, padding='<END>')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import MiniBatchKMeans
from scipy.sparse import hstack
from collections import defaultdict
```

```python
lefts_s = [' '.join([str(x) for x in l]) for l in lefts]
rights_s = [' '.join([str(x) for x in l]) for l in rights]
middles_s = [' '.join([str(x) for x in l]) for l in middles]
```

```python
tfidf = TfidfVectorizer(max_features=3000)
tfidf.fit(lefts_s + rights_s + middles_s)
```

```python
l = tfidf.fit_transform(lefts_s)
r = tfidf.fit_transform(rights_s)
m = tfidf.fit_transform(middles_s)
```

```python
X = hstack([l,m,r])
```

```python
X.shape
```

```python
cluster = MiniBatchKMeans(100, verbose=1, reassignment_ratio=0.3, max_no_improvement=5000)
cluster.fit(X)
```

```python
cl2id = defaultdict(list)

for i, cl in enumerate(cluster.labels_):
    cl2id[cl].append(i)
```

```python
f = open('clusters.txt', 'w')

for cl in cl2id:
    f.write('CLUSTER __' + str(cl) + '__\n')
    for i in cl2id[cl]:
        f.write('\t'.join([str(ents1[i]), str(ents2[i])]) + '\n')
        f.write(sents[i].replace('\n', ' ') + '\n')
    f.write('\n\n')
f.close()
```
